# **Understanding Cross-Validation and Standardization**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn import datasets
```

```python
cancer = datasets.load_breast_cancer(as_frame=True)
```

```python
cancer.target_names
```

```python
cancer.frame
```

```python
X = cancer.data
y = cancer.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 17)
```

Changing Value of k

```python
neighbors = []
scores_list = []

for k in range(1, 25):
    neighbors.append(k)
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train,y_train)
    scores = knn.score(X_test,y_test)
    scores_list.append(scores)
```

```python
Accuracy = [x*100 for x in scores_list]
# plot Accuracy versus k
plt.figure(figsize = (10, 6))
plt.plot(neighbors, Accuracy)
plt.xlim([0, 25])
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy in %')
plt.show()
```

```python
# determining the best k
max_acc = max(Accuracy)
optimal_ks = [neighbors[i] for i, j in enumerate(Accuracy) if j == max_acc]
print(max_acc)
print(optimal_ks)
```

AUC-ROC Curve

```python
## AUC-ROC
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import matplotlib.pyplot as plt

def plot_roc(y_test,predicted):
  y_scores = predicted
  fpr, tpr, threshold = roc_curve(y_test, y_scores)
  roc_auc = auc(fpr, tpr)
  plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
  plt.legend(loc = 'lower right')
  plt.plot([0, 1], [0, 1],'r--')
  plt.xlim([0, 1])
  plt.ylim([0, 1])
  plt.ylabel('True Positive Rate')
  plt.xlabel('False Positive Rate')
  plt.title('Receiver Operating Characteristic')
  plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train,y_train)
predictions = knn.predict(X_test)
plot_roc(y_test,predictions)
```

Cross Validation

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 17)
```

```python
from sklearn.model_selection import KFold
from statistics import mean
kf = KFold(n_splits=10)
cv_scores = []
neighbors = []
for k in range(1,25):
  knn = KNeighborsClassifier(n_neighbors = k)
  neighbors.append(k)
  acclist = []
  for train_index, test_index in kf.split(X_train, y_train):
    x_train_fold, x_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
    y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]
    knn.fit(x_train_fold, y_train_fold)
    acclist.append(knn.score(x_test_fold, y_test_fold))
  cv_scores.append(mean(acclist))
```

```python
Accuracy = [x*100 for x in cv_scores]

# plot misclassification error versus k
plt.figure(figsize = (10, 6))
plt.plot(neighbors, Accuracy)
plt.xlabel('Number of neighbors')
plt.ylabel('Mean Cross-validation accuracy in %')
plt.show()
```

```python
# determining the best k
max_acc = max(Accuracy)
optimal_ks = [neighbors[i] for i, j in enumerate(Accuracy) if j == max_acc]
print(optimal_ks)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from statistics import mean
import matplotlib.pyplot as plt

# Initialize Stratified K-Fold with 10 splits
skf = StratifiedKFold(n_splits=10)

# List to store cross-validation scores and k values
cv_scores = []
neighbors = []

# Loop through possible k values
for k in range(1, 25):
    knn = KNeighborsClassifier(n_neighbors=k)
    neighbors.append(k)
    acclist = []

    # Perform Stratified K-Fold Cross-Validation
    for train_index, test_index in skf.split(X_train, y_train):
        x_train_fold, x_test_fold = X_train.iloc[train_index], X_train.iloc[test_index]
        y_train_fold, y_test_fold = y_train.iloc[train_index], y_train.iloc[test_index]
        knn.fit(x_train_fold, y_train_fold)
        acclist.append(knn.score(x_test_fold, y_test_fold))

    # Calculate the mean accuracy for the current k
    cv_scores.append(mean(acclist))

# Convert cross-validation accuracy to percentage
Accuracy = [x * 100 for x in cv_scores]

# Plotting misclassification error vs. k
plt.figure(figsize=(10, 6))
plt.plot(neighbors, Accuracy, marker='o')
plt.xlabel('Number of neighbors')
plt.ylabel('Mean Cross-validation accuracy in %')
plt.title('Accuracy vs. Number of Neighbors (Stratified K-Fold)')
plt.show()

# Determine the best k
max_acc = max(Accuracy)
optimal_ks = [neighbors[i] for i, j in enumerate(Accuracy) if j == max_acc]
print("Optimal k values with Stratified K-Fold:", optimal_ks)
```

```python
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
from statistics import mean
import matplotlib.pyplot as plt

# Standardize the feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Set up K-Fold Cross-Validation
kf = KFold(n_splits=10, shuffle=True, random_state=17)

# Lists to store results
cv_scores = []
neighbors = []

# Loop over different k values
for k in range(1, 25):
    knn = KNeighborsClassifier(n_neighbors=k)
    neighbors.append(k)
    acclist = []

    # Perform K-Fold Cross-Validation
    for train_index, test_index in kf.split(X_scaled):
        x_train_fold, x_test_fold = X_scaled[train_index], X_scaled[test_index]
        y_train_fold, y_test_fold = y[train_index], y[test_index]
        knn.fit(x_train_fold, y_train_fold)
        acclist.append(knn.score(x_test_fold, y_test_fold))

    # Calculate mean accuracy for the current k
    cv_scores.append(mean(acclist))

# Convert accuracy scores to percentages
Accuracy = [x * 100 for x in cv_scores]

# Plot Accuracy vs. k
plt.figure(figsize=(10, 6))
plt.plot(neighbors, Accuracy, marker='o')
plt.xlabel('Number of neighbors')
plt.ylabel('Mean Cross-validation accuracy in %')
plt.title('Accuracy vs. Number of Neighbors (K-Fold Cross-Validation on Standardized Data)')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from scipy.spatial import Voronoi, voronoi_plot_2d

# Assuming X_scaled and y are already defined and standardized from previous steps

# Step 1: Reduce data to 2D using PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_scaled)

# Step 2: Train KNN with the optimal k from Task-3
optimal_k = neighbors[np.argmax(Accuracy)]
knn_optimal = KNeighborsClassifier(n_neighbors=optimal_k)
knn_optimal.fit(X_reduced, y)

# Step 3: Define the mesh for the decision boundary plot
x_min, x_max = X_reduced[:, 0].min() - 1, X_reduced[:, 0].max() + 1
y_min, y_max = X_reduced[:, 1].min() - 1, X_reduced[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Step 4: Predict the classes for each point in the mesh to plot decision boundaries
Z = knn_optimal.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.figure(figsize=(10, 6))
cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#0000FF'])
plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.4)

# Plot the training points
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=20)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title(f'KNN Decision Boundary with k={optimal_k} (2D PCA Projection)')

# Step 5: Plot the Voronoi Diagram
vor = Voronoi(X_reduced)
voronoi_plot_2d(vor, ax=plt.gca(), show_vertices=False, line_colors='black', line_width=0.5, line_alpha=0.6)
```

**Questions**

**Instead of using cross-validation, we could also split our dataset into three sets- train, validation, and test set. The training proceeds on the training set, after which evaluation is done on the validation set, and when the experiment seems to be successful, final evaluation can be done on the test set.**

**1. What do you think are the pros/cons of this method ?**


**Pros:**

Simple to Implement: Splitting data into three sets is straightforward and requires less computational effort than repeated cross-validation. \\
Clear Division of Data Roles: Each dataset (training, validation, and test) has a distinct purpose, reducing the risk of overfitting if handled properly. \\
Less Computational Cost: With only one training cycle and validation, this approach is faster and less computationally expensive, which can be advantageous for very large datasets or complex models.

**Cons:**

Less Reliable for Small Datasets: When the dataset is small, dividing it into three parts can lead to unreliable performance estimates. Each split has less data, which increases variance in performance metrics. \\
Higher Variability in Results: The performance can vary significantly depending on the specific split, meaning the choice of training, validation, and test samples may affect the model's reported accuracy. \\
Wasted Data: Only a portion of the data is used for training, leading to potentially under-utilizing the available data, especially when training data is limited.

**2. Is K-Fold cross validation an exhaustive or non-exhaustive cross validation method ? Justify your answer.**

K-Fold Cross-Validation is a non-exhaustive cross-validation method. In K-Fold Cross-Validation, the dataset is split into k subsets, and the model is trained and validated k times, each time using a different fold as the validation set. However, it doesnâ€™t use every possible subset or combination of training and validation splits. An exhaustive cross-validation method would include Leave-One-Out Cross-Validation (LOOCV), where every possible training/validation split is used. In LOOCV, each observation in the dataset is used once as a validation instance, with the remaining data used for training, iterating over all data points. This is exhaustive because it covers all possible ways to split the data with one observation left out each time.
