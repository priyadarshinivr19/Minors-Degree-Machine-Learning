# **Basic Plots**

```python
import requests
url = "https://raw.githubusercontent.com/SP9144/FMML_Data/main/Automobile_data.csv"
response = requests.get(url)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from io import StringIO
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
data = StringIO(response.text)
data = pd.read_csv(data)
print("Number of data points =", len(data))
data.head()
```

```python
# Removing rows which have missing values

data.replace("?", np.nan, inplace = True)
data = data.dropna()
print(len(data))
```

```python
# Dividing the dataset into features (X) and the target variable (y)

X = data.drop("price", axis = 1)
y = pd.to_numeric(data["price"])
```

```python
plt.figure(figsize = (24, 7))
plt.scatter(X["make"], y)
plt.xlabel('Make')
plt.ylabel('Price')
plt.title('Car maker vs Price - Scatter Plot')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(24,7)})
sns.boxplot(x=X["make"],y=y, palette="Set3").set_title('Car Manufacturer vs Price - Box Plot')
```

```python
sns.violinplot(x=X["make"],y=y, palette="Set3").set_title('Car maker vs Price - Violin Plot')
```

```python
sns.swarmplot(x=X["make"],y=y).set_title('Car maker vs Price Swarm Plot')
```

```python
sns.boxplot(x=X["make"],y=pd.to_numeric(X["horsepower"]), palette="Set3").set_title('Car maker vs Horsepower Box Plot')
```

```python
sns.set(rc={'figure.figsize':(7,5)})
sns.scatterplot(x=pd.to_numeric(X["horsepower"]), y = y, color = "red")
```

```python
sns.jointplot(x=pd.to_numeric(X["horsepower"]), y = y, kind="reg", color = 'green')
```

# **Exercise 1**

Plot the jointplot for a feature which has negative correlation with the target variable "price".

```python
# List of features
features = ['symboling', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size',
            'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg']

# Generate joint plots for each feature
for feature in features:
    sns.jointplot(x=pd.to_numeric(data[feature]), y=y, kind="reg", color='purple')
    plt.title(f'Relationship between {feature} and price')
    plt.show()
```

Symboling, peak rom, city-mpg and highway-mpg have negative correlation with price.

1. Plotting multiple plots together: Search for information about plt.subplots(). Choose any 2 plot types from the ones discussed above and plot them together in a single figure using subplots.

```python
# Creating subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# First subplot: Regression plot of 'wheel-base' vs 'price'
sns.regplot(x=pd.to_numeric(data['wheel-base']), y=y, ax=axes[0], color='purple')
axes[0].set_title('Wheel-base vs Price')

# Second subplot: Scatter plot of 'horsepower' vs 'price'
sns.scatterplot(x=pd.to_numeric(data['horsepower']), y=y, ax=axes[1], color='green')
axes[1].set_title('Horsepower vs Price')

plt.tight_layout()
plt.show()
```
